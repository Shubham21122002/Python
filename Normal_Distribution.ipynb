```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Step 1: Generate random test results (assuming scores between 0 and 100)
np.random.seed(42)
test_scores = np.random.normal(loc=70, scale=10, size=100)  # Mean=70, StdDev=10

# Step 2: Plot the distribution of test scores
plt.figure(figsize=(8, 6))
sns.histplot(test_scores, kde=True, color='blue', bins=15, stat='density')
plt.title("Distribution of Class Test Scores")
plt.xlabel("Test Scores")
plt.ylabel("Density")

# Step 3: Calculate and display Skewness and Kurtosis
skewness = stats.skew(test_scores)
kurtosis = stats.kurtosis(test_scores)

print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")

# Step 4: Plot the normal distribution curve for reference
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, np.mean(test_scores), np.std(test_scores))
plt.plot(x, p, 'k', linewidth=2, label="Normal Distribution")
plt.legend()

# Show the plot
plt.show()
```
