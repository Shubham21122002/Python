```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data ,hue='species', palette='Set1')
```

```python
sns.displot(data, x="sepal_length", kind="kde")
plt.show()
```

```python
sns.displot(data, x="sepal_length", kind="kde")
```

```python
import matplotlib.pyplot as plt
palette = sns.color_palette('Blues',11)
sns.palplot(palette)
plt.show()
```

```python
palette = sns.color_palette('Greens',11)
sns.palplot(palette)
plt.show()
```

```python
sns.barplot(x='species', y='sepal_length', data=data , hue='species', palette='Set2')
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_width', data=data , hue='species', palette='Set2')
plt.show()
```

```python
sns.violinplot(x='species', y='sepal_width', data=data, hue='species', palette='Set2')
plt.show()
```

```python
sns.histplot(x='species', y='sepal_width', data=data , hue='species', palette='Set2')
plt.show()
```
