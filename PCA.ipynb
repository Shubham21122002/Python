```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns

# Step 2: Create a simple dataset (let's simulate some data)
np.random.seed(42)
data = pd.DataFrame({
    'Feature1': np.random.normal(10, 2, 100),
    'Feature2': np.random.normal(20, 5, 100),
    'Feature3': np.random.normal(30, 3, 100)
})

# Step 3: Check data correlation
correlation_matrix = data.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Step 4: Rescaling the data (Standardization)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 5: Applying PCA for Dimensionality Reduction
pca = PCA(n_components=2)  # Reduce data to 2 dimensions
pca_data = pca.fit_transform(scaled_data)

# Step 6: Print explained variance (how much information we retain after reduction)
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance by each PCA component: {explained_variance}")

# Step 7: Visualizing the PCA result
plt.figure(figsize=(6,6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c='blue', label="PCA Result")
plt.title('PCA: 2D Projection of the Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
